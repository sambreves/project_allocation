```python
import pandas as pd

df = pd.read_csv('.././data/files/log_cotacoes.csv', dtype=str, sep=';')
df2 = pd.read_csv('.././data/files/variations_produtos_v2.csv')

df['Produto Descrição\nIntegradora'] = df['Produto Descrição\nIntegradora'].str.strip()

SKU = [200102, 200104, 200105, 200110, 200111, 200112, 200114, 200115, 200124, 200125, 200130, 200131, 200132, 200134, 200135, 200165, 200154, 200164, 200181]

result = pd.DataFrame()

for sku in SKU:
    list_variations_products = df2.loc[(df2['SKU']==sku), 'Variacao'].values.tolist()

    variations_produtos_text = '|'.join(list_variations_products)

    df_filtrado = df.loc[df['Produto Descrição\nIntegradora'].str.contains(variations_produtos_text, case=False, na=False), :]
    df_filtrado['SKU'] = sku

    if result.empty:
        result = df_filtrado
    else:
        result = pd.concat([result, df_filtrado], axis=0, ignore_index=True)

```
