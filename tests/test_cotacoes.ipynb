```python
import pandas as pd

df = pd.read_csv('.././data/files/log_cotacoes.csv', dtype=str, sep=';')
```

```python
df.columns = df.columns.str.strip()
```

```python
variations_produtos = pd.read_csv('.././data/files/variations_produtos.csv')
variations_produtos.head()
```

```python
list_variations_products = variations_produtos['Variacao'].values.tolist()
list_variations_products
```

```python
variations_produtos_text = '|'.join(list_variations_products)
variations_produtos_text
```

```python
df_filtrado = df[df['Produto Descrição\nIntegradora'].str.contains(variations_produtos_text, case=False, na=False)]
```

```python
df_filtrado.to_excel('.././data/files/log_cotacoes_filtrado.xlsx', index=False)
df_filtrado.head()
```

```python
import re
import numpy as np

def normalize_text(text):
    """
    Função para limpar e padronizar o texto, tornando a comparação mais eficaz.
    """
    if not isinstance(text, str):
        return ""
    
    text = text.lower()  # 1. Converte para minúsculas
    text = text.replace(',', '.')  # 2. Troca vírgula por ponto
    text = re.sub(r'(\d)([a-zA-Z%])', r'\1 \2', text)  # 3. Separa números de letras/unidades (ex: 100ml -> 100 ml)
    text = re.sub(r'[^\w\s.]', ' ', text)  # 4. Remove toda pontuação exceto pontos (de decimais)
    text = re.sub(r'\s+', ' ', text).strip()  # 5. Remove espaços extras
    return text

try:
    # Carrega os arquivos.
    # O 'log_cotacoes_filtrado.csv' usa ponto e vírgula (;) como separador.
    log_df = df_filtrado
    keywords_df = variations_produtos

    # --- Lógica Final: Algoritmo de Melhor Correspondência com Limite Flexível ---

    # 1. Preparação: Normaliza e pré-processa a lista de palavras-chave
    keywords_df.dropna(subset=['Variacao'], inplace=True)
    
    keyword_list = []
    for keyword, sku in zip(keywords_df['Variacao'], keywords_df['SKU']):
        normalized_keyword = normalize_text(keyword)
        if normalized_keyword:
            keyword_parts = normalized_keyword.split()
            if keyword_parts:
                keyword_list.append((keyword_parts, sku))

    # 2. Função principal para encontrar o SKU baseado na melhor pontuação
    def encontrar_melhor_sku(descricao):
        descricao_normalizada = normalize_text(descricao)
        if not descricao_normalizada:
            return 0

        best_score = 0
        best_sku = 0
        best_match_count = 0
        len_best_keyword = 0

        for keyword_parts, sku in keyword_list:
            # Calcula quantas partes da keyword estão na descrição
            match_count = sum(1 for part in keyword_parts if part in descricao_normalizada)
            
            # Calcula a pontuação (0.0 a 1.0)
            score = match_count / len(keyword_parts)

            # Regra de atualização:
            # - Se a pontuação é maior, temos um novo vencedor.
            # - Se a pontuação é a mesma, o vencedor é a keyword mais longa (mais específica).
            if score > best_score:
                best_score = score
                best_sku = sku
                best_match_count = match_count
                len_best_keyword = len(keyword_parts)
            elif score == best_score and len(keyword_parts) > len_best_keyword:
                best_sku = sku
                best_match_count = match_count
                len_best_keyword = len(keyword_parts)

        # Decisão Final: Aceita o SKU se a pontuação for boa o suficiente.
        # Condições: Pontuação >= 75% E pelo menos 2 palavras correspondendo.
        if best_score >= 0.75 and best_match_count >= 2:
            return best_sku
        else:
            return 0

    # 3. Aplica a função final para criar a coluna SKU
    print("Iniciando a busca por SKUs com a lógica de melhor correspondência...")
    # Garante que a coluna de descrição seja do tipo string antes de aplicar a função
    log_df['SKU'] = log_df['Produto Descrição\nIntegradora'].astype(str).apply(encontrar_melhor_sku)
    print("Busca finalizada.")
    
    # --- Fim da Lógica ---

    log_df['SKU'] = log_df['SKU'].fillna(0)

    output_filename = '.././data/files/log_cotacoes_tratado.xlsx'
    log_df.to_excel(output_filename, index=False)

    print(f"\nProcesso concluído!")
    print(f"O resultado foi salvo no arquivo: {output_filename}")
    
    print("\nVisualização das primeiras 20 linhas do resultado:")
    print(log_df[['Produto Descrição\nIntegradora', 'SKU']].head(20))
    
    skus_encontrados = log_df[log_df['SKU'] != 0].shape[0]
    total_linhas = log_df.shape[0]
    if total_linhas > 0:
        percentual = (skus_encontrados / total_linhas) * 100
        print(f"\nResumo: {skus_encontrados} de {total_linhas} linhas receberam um SKU ({percentual:.2f}%).")


except FileNotFoundError as e:
    print(f"\nErro: Arquivo não encontrado. Verifique se os nomes dos arquivos estão corretos.")
    print(f"Detalhes: {e}")
except Exception as e:
    print(f"\nOcorreu um erro inesperado: {e}")
```
